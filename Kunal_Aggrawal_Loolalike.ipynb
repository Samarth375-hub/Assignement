```python
product_df = pd.read_csv('products.csv')
```

```python
final_df = pd.merge(transaction_df, customer_df, on='CustomerID', how='inner')
final_df = pd.merge(final_df, product_df, on=['ProductID','Price'], how='inner')
```

```python
final_df
```

```python
final_df.drop(columns = ['CustomerName','ProductName'],axis = 1,inplace = True)
```

```python
final_df
```

```python
final_df['TransactionDate'] = pd.to_datetime(final_df['TransactionDate'])
final_df['SignupDate'] = pd.to_datetime(final_df['SignupDate'])
```

```python
final_df
```

```python
final_df['TransactionDate'].dt.day.max()
```

```python
time = pd.Timestamp('2024-12-31')
Recent = final_df.groupby('CustomerID')['TransactionDate'].max().apply(lambda x: (time-x).days)
```

```python
Recent
```

```python
frequency = final_df.groupby('CustomerID')['TransactionID'].count()
```

```python
frequency
```

```python
total_spend = final_df.groupby('CustomerID')['TotalValue'].sum()
```

```python
total_spend
```

```python
category = final_df.groupby(['CustomerID','Category']).size().unstack(fill_value = 0)
```

```python
category
```

```python
final_features = pd.DataFrame({'Recent':Recent,'Frequency': frequency,'Total_spend':total_spend}).join(category).reset_index()
```

```python
final_features
```

```python
from sklearn.preprocessing import StandardScaler
features = final_features.drop(columns = ['CustomerID'])
scaler = StandardScaler()
standardized_features = scaler.fit_transform(features)
```

```python
features
```

```python
standardized_df = pd.DataFrame(standardized_features,columns = features.columns,index = final_features['CustomerID'])
```

```python
standardized_df
```

```python
from sklearn.neighbors import NearestNeighbors

knn = NearestNeighbors(n_neighbors = 4,metric = 'cosine')
knn.fit(standardized_df)
```

```python
distances,indices = knn.kneighbors(standardized_df)
```

```python
distances
```

```python
customer_ids = standardized_df.index
lookalikes = {}

top_customers = [f'C{i:04d}' for i in range(1,21)]

for idx, customer_id in enumerate(customer_ids):
    if customer_id in top_customers:  
        similar_customers = customer_ids[indices[idx][1:4]]   
        similar_scores = 1 - distances[idx][1:4]  
       
       
        filtered_lookalikes = [(cust, score) for cust, score in zip(similar_customers, similar_scores)]
      
        lookalikes[customer_id] = filtered_lookalikes[:3]  
       
```

```python
lookalike_df = pd.DataFrame({
    'CustomerID':lookalikes.keys(),
    'Lookalikes':[str(v) for v in lookalikes.values()]
})


```

```python
lookalike_df.to_csv('Kunal_Aggrawal_Lookalike.csv',index  = False)
```
